## Modeling

For the score evaluation of the different models i will use the previously showed function provided by the course.


```python
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix,ConfusionMatrixDisplay, precision_recall_fscore_support, precision_score, recall_score
def evaluate_metrics(yt, yp):
    results_pos = {}
    results_pos['accuracy'] = accuracy_score(yt, yp)
    precision, recall, f_beta, _ = precision_recall_fscore_support(yt, yp)
    results_pos['recall'] = recall
    results_pos['precision'] = precision
    results_pos['f1score'] = f_beta
    return results_pos
```

My first prediction model will be Logistic regresion as we can easily review the weights of the coeficients and get some valuable insights.

### Train-Test split

```python
# Random state
rs = 502
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state = rs)
```

### Resampling

```python
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import SMOTE

def resample(X_train, y_train):
    # SMOTE sampler (Oversampling)
    smote_sampler = SMOTE(random_state = 123)
    # Undersampling
    under_sampler = RandomUnderSampler(random_state=123)
    # Resampled datasets
    X_smo, y_smo = smote_sampler.fit_resample(X_train, y_train)
    X_under, y_under = under_sampler.fit_resample(X_train, y_train)
    return X_smo, y_smo, X_under, y_under
```

```python
X_smo, y_smo, X_under, y_under = resample(X_train, y_train)
```

### Logistic Regression

Now i will fit and predict a logistic regression with the original dataset, the dataset with SMOTE, and the dataset wiht UnderSampling.  
Note that i am using the balanced class_weight in the regression with the original dataset which automatically adjusts weights inversely proportional to class frequencies to counter the imbalance of the dataset.  

```python
from sklearn.linear_model import LogisticRegression
Lr = LogisticRegression(max_iter=1000, random_state = rs, class_weight='balanced')
# Predictions with the original dataset
Lr.fit(X_train, y_train)
Lr_pred = Lr.predict(X_test)

Lr = LogisticRegression(max_iter=1000, random_state = rs)
# Predictions after SMOTE resampling
Lr.fit(X_smo, y_smo)
Lr_SMOTE_pred = Lr.predict(X_test) 

# Predictions after UnderSampling
Lr.fit(X_under, y_under)
Lr_under_pred = Lr.predict(X_test)

scores = pd.DataFrame({
    'LogisticRegression': evaluate_metrics(y_test, Lr_pred),
    'LogisticRegression_SMOTE': evaluate_metrics(y_test, Lr_SMOTE_pred),
    'LogisticRegression_UnderSampling': evaluate_metrics(y_test, Lr_under_pred)
})

```

### KNN

Following the logistic regression i will do the same with a K-Nearest Neighbors

For starters i will do some hyperparameter tuning to find the optimal number of neighbors to use  with the same resampled dataset as well as the original one for them to perform to their best

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier

param_grid = {'n_neighbors': range(1, 100 + 1)}
knn = KNeighborsClassifier()
grid_search = GridSearchCV(estimator=knn, param_grid=param_grid, cv=5)

grid_search.fit(X_train, y_train.values.ravel())
best_k = grid_search.best_params_['n_neighbors']
scores['KNN'] = evaluate_metrics(y_test, grid_search.best_estimator_.predict(X_test))

grid_search.fit(X_smo, y_smo.values.ravel())
best_k_smo = grid_search.best_params_['n_neighbors']
scores['KNN_SMOTE'] = evaluate_metrics(y_test, grid_search.best_estimator_.predict(X_test))

grid_search.fit(X_under, y_under.values.ravel())
best_k_under = grid_search.best_params_['n_neighbors']
scores['KNN_UnderSampling'] = evaluate_metrics(y_test, grid_search.best_estimator_.predict(X_test))

```

```python
print("Best k:",best_k)
print("Best k for SMOTE:",best_k_smo)
print("Best k eith UnderSampling:",best_k_under)
```

### Gradient Boosting Classifier

This model was chosen due to the imbalance of the dataframe since it has the posibility of fine tuning several paramters to obtain better results and with the way it works learning from previous mistakes i hope it will be better at predicting the minority class.

The same as KNN i will start with some hyperparameter tuning.

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import uniform

param_grid = {
    'learning_rate': uniform(0.01, 0.3),
    'n_estimators': [200, 300, 400],
    'max_depth': [3, 5, 6],
    'subsample': [0.8, 1],
    'min_samples_split': [2, 5, 10]
}

Gb = GradientBoostingClassifier(random_state=rs)
random_search = RandomizedSearchCV(
    estimator=Gb, 
    param_distributions=param_grid, 
    n_iter=50, 
    cv=2, 
    verbose=1,
    n_jobs=-1,
    random_state=rs)

random_search.fit(X_train, y_train.values.ravel())
best_params = random_search.best_params_

scores['Gradient_Boosting'] = evaluate_metrics(y_test, random_search.predict(X_test))

best_gb = GradientBoostingClassifier(random_state=rs, **best_params)

best_gb.fit(X_smo, y_smo.values.ravel())
scores['Gradient_Boosting_SMOTE'] = evaluate_metrics(y_test, best_gb.predict(X_test))

best_gb.fit(X_under, y_under.values.ravel())
scores['Gradient_Boosting_UnderSampling'] = evaluate_metrics(y_test, best_gb.predict(X_test))
```

## Insights

### Logistic Regression

For logistic regression we can see that even though it has an overall acurracy of aproximately 73% the minority class is practically ignored in most scores this being the one that we are trying to predict.  
Because of this i wouldnt recommend using logistic regression for this case study or with any dataset with extreme imbalance.  
Since we are already compensating for the imbalance with the 'balanced' class weight option there is not much change with SMOTE or UnderSampling.  

### KNN

Knn performed significantly better in  acurracy all though it completely ignores the minority class makeing this model useless as is, when compared with the model trained in the resampled datasets we can se that even though the overall acurracy is lower there is at least some precision with the minority class and in the case of KNN with SMOTE there is a way better balance betwen scores and still keeping an acceptable acurracy that not being the case with under sampling.   

### Gradient Boosting

As for gradient boosting it is by far the best method withount any resampling  and with resamplig it is still better balanced across the board.  

## Recomendations

Best Overall Model: Gradient Boosting with SMOTE

High accuracy, improved recall, and better precision balance make it the most robust choice across metrics.
Suitable when both classes need reasonable performance, even if some precision for the minority class is sacrificed.
When Minority Class Recall Is Critical:

Consider Gradient Boosting with Undersampling, as it achieves the best minority recall (~0.95). However, the trade-offs in precision and overall accuracy must be acceptable.
When Precision for Both Classes Matters:

KNN with SMOTE or Logistic Regression with SMOTE may be better options since they offer moderate precision and less severe trade-offs than undersampling.
